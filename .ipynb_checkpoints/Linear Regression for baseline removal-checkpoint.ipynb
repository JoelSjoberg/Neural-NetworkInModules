```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
spectral_line = [0.2, 0.3, 4.2, 5.4, 6.3, 4.3, 3.5, 0.1, 0.3, 0.2, 4, 29.3, 549, 90.2, 3.4, 0.3, 0.1, 0.6, 0.3, 0.4]
plt.plot(spectral_line)
plt.show()
```

```python
pol_order = 1
pol_order = pol_order + 1
params = np.array([np.random.random_sample() - 0.5 for p in range(pol_order)])
bias = np.random.random_sample()

print("Params:", params)

def lin_reg(x, params):
    pred = 0
    for i in range(pol_order):
        pred += params[i] * np.power(x, i)
    return pred

def update(x, true_val, params, alpha = 0.003):
    error = (lin_reg(x, params) - true_val) * 2
    
    for i in range(pol_order):
        params[i] -= alpha * error * np.power(x, i)
    
    return params

def train_on_points(points, params, epoch = 100):
    param_hist = []
    for i in range(epoch):
        for p in range(len(points)):

            params = update(p, points[p])

        param_hist.append(params)
    
    param_hist = np.array(param_hist)
    print(param_hist)
    plt.plot(param_hist.T[0])
    plt.plot(param_hist.T[1])
    plt.show()

train_on_points(spectral_line)
```

```python
x = np.linspace(0, len(spectral_line), num = len(spectral_line))
list(map(lin_reg, x))
```

```python
x = np.linspace(0, len(spectral_line), num = len(spectral_line))
plt.plot(x, list(map(lin_reg, x)))
plt.show()
```
